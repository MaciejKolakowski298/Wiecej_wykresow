```python
prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),

]
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(prices)
```

```python
df
```

```python
df=df.reset_index()
```

```python
df
```

```python
df=df.rename({0:'miesiąc', 1:'cena_PLN'}, axis=1)
```

```python
df
```

```python
df=df.set_index('miesiąc')
df=df.drop('index', axis=1)
```

```python
df
```

```python
df['cena_USD']=df['cena_PLN']*4
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
figure, ax = plt.subplots()
ax.plot(df['cena_USD'], color='red', linestyle='dashed')
ax.set_xlabel("miesiąc")
ax.set_ylabel("cena w USD")
```
